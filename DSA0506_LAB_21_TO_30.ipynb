```python
# First create test.txt file
with open('test.txt', 'w') as f:
    f.write("1 2\n2 4\n3 1")

# Read and plot
x, y = [], []
with open('test.txt', 'r') as f:
    for line in f:
        a, b = map(float, line.split())
        x.append(a)
        y.append(b)

plt.plot(x, y, marker='o')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Line from File Data')
plt.show()
```

```python
import pandas as pd
data = {
    'Date': ['10-03-16', '10-04-16', '10-05-16', '10-06-16', '10-07-16'],
    'Close': [772.56, 776.43, 776.47, 776.86, 775.08]
}
df = pd.DataFrame(data)
plt.plot(df['Date'], df['Close'], marker='o')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Alphabet Stock (Oct 3-7, 2016)')
plt.xticks(rotation=45)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.plot(x, y1, label='Line 1', color='red', linewidth=3)
plt.plot(x, y2, label='Line 2', color='blue', linewidth=1)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Multiple Lines')
plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1
axs[0,0].plot([1,2,3], [1,4,9])
axs[0,0].set_title('Plot 1')

# Plot 2
axs[0,1].bar(['A','B','C'], [3,7,2])
axs[0,1].set_title('Plot 2')

# Plot 3
axs[1,0].scatter([1,2,3,4], [1,4,9,16])
axs[1,0].set_title('Plot 3')

# Plot 4
axs[1,1].pie([30,20,50], labels=['A','B','C'])
axs[1,1].set_title('Plot 4')

plt.tight_layout()
plt.show()
```

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.bar(languages, popularity)
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Programming Language Popularity')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.barh(languages, popularity)
plt.xlabel('Popularity (%)')
plt.ylabel('Programming Languages')
plt.title('Programming Language Popularity (Horizontal)')
plt.show()
```

```python
colors = ['red', 'green', 'blue', 'orange', 'purple', 'brown']
plt.bar(languages, popularity, color=colors)
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Colored Bar Chart')
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np
groups = ['Group1', 'Group2', 'Group3', 'Group4', 'Group5']
men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]

x = np.arange(len(groups))
width = 0.35

plt.bar(x - width/2, men, width, label='Men', color='blue')
plt.bar(x + width/2, women, width, label='Women', color='pink')

plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title('Scores by Group and Gender')
plt.xticks(x, groups)
plt.legend()
plt.show()
```
