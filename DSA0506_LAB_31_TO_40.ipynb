```python
import matplotlib.pyplot as plt
import numpy as np
men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]
men_std = [4, 3, 4, 1, 5]
women_std = [3, 5, 2, 3, 3]
groups = ['G1', 'G2', 'G3', 'G4', 'G5']
x = np.arange(len(groups))
plt.bar(x, men, yerr=men_std, label='Men')
plt.bar(x, women, yerr=women_std, bottom=men, label='Women')
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title('Stacked Bar with Error Bars')
plt.xticks(x, groups)
plt.legend()
plt.show()
```

```python
np.random.seed(42)
x = np.random.randn(50)
y = np.random.randn(50)

plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random Scatter Plot')
plt.grid(True)
plt.show()
```

```python
plt.scatter(x, y, facecolors='none', edgecolors='blue', s=100)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot (Empty Circles)')
plt.grid(True)
plt.show()
```

```python
sizes = np.random.uniform(20, 500, 50)
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.7)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Different Sizes')
plt.colorbar(label='Color intensity')
plt.show()
```

```python
math_marks = [88, 92, 80, 89, 100, 80, 60, 100, 80, 34]
science_marks = [35, 79, 79, 48, 100, 88, 32, 45, 20, 30]

plt.scatter(range(10), math_marks, label='Math', color='red', s=100)
plt.scatter(range(10), science_marks, label='Science', color='blue', s=100)
plt.xlabel('Student Index')
plt.ylabel('Marks')
plt.title('Math vs Science Marks')
plt.legend()
plt.show()
```

```python
# EXPERIMENT 36
np.random.seed(42)
heights = np.concatenate([np.random.normal(160, 10, 20),
                          np.random.normal(170, 8, 20),
                          np.random.normal(165, 12, 20)])
weights = np.concatenate([np.random.normal(60, 8, 20),
                          np.random.normal(70, 10, 20),
                          np.random.normal(65, 9, 20)])
groups = ['Group A']*20 + ['Group B']*20 + ['Group C']*20

for group in ['Group A', 'Group B', 'Group C']:
    mask = [g == group for g in groups]
    plt.scatter([h for h, m in zip(heights, mask) if m],
                [w for w, m in zip(weights, mask) if m],
                label=group, s=50, alpha=0.7)

plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Weight vs Height (3 Groups)')
plt.legend()
plt.grid(True)
plt.show()
```
